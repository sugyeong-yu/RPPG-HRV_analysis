```python
subject=0# 0~10
path="D:\\prlab\\ysg\\rppg\\RPPG-HRV_analysis\\result\\time\\normalized\\%d.csv"%(subject)
data=pd.read_csv(path)

# c_lf_hf_ratio=data['cppg_lf_hf_ratio']
# c_lfnu=data['cppg_lfnu']
# c_hfnu=data['cppg_hfnu']

# r_lf_hf_ratio=data['rppg_lf_hf_ratio']
# r_lfnu=data['rppg_lfnu']
# r_hfnu=data['rppg_hfnu']
# print(len(r_hfnu))

c_meannni=data['cppg_mean_nn']
r_meannni=data['rppg_mean_nn']

c_sdnn=data['cppg_sdnn']
r_sdnn=data['rppg_sdnn']
print(len(r_sdnn))
```

```python
# NNI 일치율 구하기
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
import numpy as np
r=r_sdnn.to_numpy().reshape([-1,1])
c=c_sdnn.to_numpy().reshape([-1,1])

trainx,testx,trainy,testy=train_test_split(r[:],c[:],test_size=0.3,shuffle=True,random_state=52)

model=LinearRegression()
model.fit(trainx,trainy)
y_pred=model.predict(testx)
y_true=testy
print("R score-" ,r2_score(y_true,y_pred))

#plt.figure(figsize=(5,5))
plt.title("SDNN",fontsize=20)
plt.text(min(testx),max(testy)-3, 'R2 = {0:0.2f}'.format(r2_score(y_true,y_pred)), fontsize=20, color='black')
plt.xlabel('CPPG ')
plt.ylabel('RPPG ')
plt.rc('axes', labelsize=20)   # x,y축 label 폰트 크기
plt.rc('xtick', labelsize=15)  # x축 눈금 폰트 크기 
plt.rc('ytick', labelsize=15)
plt.scatter(testx,testy,color='red')
plt.grid(True)
plt.plot(testx,y_pred,color='gray')
#plt.savefig('ratior2')
plt.show()
```
